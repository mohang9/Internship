5. Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame-
    
i) Headline

ii) Time

iii) News Link

```python
!pip install bs4
!pip install requests
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get(' https://www.cnbc.com/world/?region=world')
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
```

```python
news_data = []

for i in soup.find_all('li', class_="LatestNews-item"):
    news_data.append(i.text)
```

```python
news_data
```

```python
time = []

for i in soup.find_all('time',class_="LatestNews-timestamp"):
    time.append(i.text)
```

```python
time
```

```python
url = []

for i in soup.find_all('a',href=True):
    url.append(i['href'])
```

```python
url
```

```python
df = pd.DataFrame({'Headlines':news_data},{'Time':time})
```
